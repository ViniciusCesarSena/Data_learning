# Média, Mediana e Moda

**Média**: A soma de todos os valores dividida pelo número de valores, representa o valor central da distribuição.

`.mean()`

---


**Mediana**: O valor que divide a lista ordenada em dois grupos com o mesmo número de elementos. Indica o "meio" da distribuição, não sendo influenciada por valores extremos.

`.median()`

---


**Moda**: O valor que aparece com mais frequência.

`.mode()`


```python
#Valores maximos e minimos são autoexplicativos
valor_minimo= df1['Item_price'].min()
valor_maximo= df1['Item_price'].max()


media= df1['Item_price'].mean()
mediana= df1['Item_price'].median()
moda= df1['Item_price'].mode()
variancia= df1['Item_price'].var()

print(f' preço minimo: {valor_minimo},\n preço maximo:  {valor_maximo}')
print(f' media: {round(media,2)},\n mediana: {mediana}')
print(f' moda: {moda}')
```

# Variancia, Desvio Padrão e Quartis

**Variancia**:

A média dos quadrados das diferenças entre cada valor e a média, mede a dispersão dos dados em torno da média. O quão "espalhados" os dados estão em relação à média, captura a ideia da "distância média" de cada valor individual em relação ao valor central do conjunto de dados.

`.var( )`

---


**Desvio** **Padrão**:

A raiz quadrada da variância, indica a amplitude da dispersão dos dados.

O desvio padrão é útil para:

**Comparar conjuntos de dado**s: Se dois conjuntos de dados possuem a mesma média, o conjunto com o maior desvio padrão terá maior dispersão dos dados.

**Avaliar a confiabilidade da média**: Um desvio padrão alto indica que a média pode não ser um bom representante dos dados, pois muitos valores estão distantes dela.

**Identificar outliers**: Valores que se distanciam em mais de 3 desvios padrões da média podem ser considerados outliers.


(Outliers são valores que se distanciam significativamente do restante dos dados. Eles podem ser causados por erros de medição, valores atípicos ou características específicas da população. )

`.std( )`

---


**Quartis**:

Os quartis são valores que dividem um conjunto de dados ordenado em quatro partes iguais. Eles são úteis para entender a distribuição dos dados e identificar possíveis assimetrias. Os quartis fornecem uma visão geral da distribuição dos dados, indicando onde estão concentrados a maioria dos valores e se há assimetria.

*   Q1: 25% dos dados - Divide os dados em 4 partes iguais = `quantile.(0.25)`
*   Q2: 50% percentil - Mediana = `quantile.(0.5)`
*   Q3: 75% dos dados = `quantile.(0.75)`

Fornecem informações sobre a distribuição dos dados.


---




**Interquartil ou IQR**:

Diferença entre o Q3 e o Q1, mede a variabilidade central dos dados, excluindo outliers.  Ele é especialmente útil quando o conjunto de dados contém muitos outliers.

O IQR é menos influenciado por outliers do que outras medidas de dispersão, como a variância e o desvio padrão. Isso porque o IQR se concentra na metade central dos dados, ignorando os valores extremos.






```python


variancia = df1['Item_price'].var()
desvio_padrao = df1['Item_price'].std()
Q1 = df1['Item_price'].quantile(0.25)
Q2 = df1['Item_price'].quantile(0.5)
Q3 = df1['Item_price'].quantile(0.75)
IQR = Q3 - Q1


print(f' variância: {round(variancia,2)}')
print(f' Desvio Padrão: {round(desvio_padrao,2)}')
print(f' Primeiro Quartil: {Q1}')
print(f' Segundo Quartil (mediana): {Q2}')
print(f' Terceiro Quartil: {Q3}')
print(f' Intervalo interquartil: {IQR} ')
```

# Assimetria e Curtose

**Assimetria(Skewness):**

A assimetria diz o quanto a distribuição dos dados está desbalanceada em torno da média. Indica se a maioria dos dados está concentrada em um lado da média ou se está distribuída de forma mais uniforme.

`.skew()`



---


Curtose:

A curtose é uma medida que complementa a assimetria, fornecendo informações sobre a forma das caudas da distribuição dos dados. Ela indica se as caudas são mais finas ou mais grossas do que em uma distribuição normal.

Curtose positiva (leptocúrtica): A curva é mais alta e fina do que a normal, com caudas mais longas e finas.

Curtose negativa (platicúrtica): A curva é mais baixa e achatada do que a normal, com caudas mais curtas e grossas.

Curtose zero (mesocúrtica): A curva é simétrica e em forma de sino, com caudas de tamanho normal.

A curtose pode influenciar a confiabilidade de intervalos de confiança, que são usados para estimar o valor real de uma população a partir de uma amostra.
Interpretação

(**As caudas** de uma distribuição de dados referem-se às extremidades da curva, onde os valores mais raros se concentram.)

`.kurtosis()`


```python


assimetria = df1['Item_price'].skew()
curtose = df1['Item_price'].kurtosis()


print(f'Assimetria(Skewness): {round(assimetria,2)} ')
print(f'Curtose: {round(curtose,2)} ')
```

#Verificar o total de valores únicos de cada variável

O código itera pelas colunas do DataFrame df1 e calcula o número de valores únicos em cada coluna após convertê-las para strings. Em seguida, ele armazena essas contagens em uma lista chamada valores_unicos.


1.   Selecionando as primeiras 21 colunas (indexação baseada em 0) de `df1.columns`
2.   Convertendo as colunas selecionadas em uma lista usando `.tolist().`
3.   Percorrendo cada nome de coluna na lista usando o loop `for`

**`.columns`**: Este atributo do DataFrame acessa os nomes das colunas (rótulos) como uma sequência (como uma lista ou array).



---

# **Primeira parte do FOR :**


1. `print(i, ':', len(df1[i].astype(str).value_counts()))`

`i`: Representa o nome da coluna atual sendo iterada no loop (assumindo que está dentro de um loop iterando sobre df1.columns).

`df1[i]`: Seleciona uma coluna específica do DataFrame df_dados com base no nome da coluna i.

`.astype(str)`: Converte os dados na coluna selecionada para strings. Isso é necessário porque value_counts() geralmente funciona melhor com dados categóricos (strings ou categorias numéricas).

`.value_counts()`: Conta as ocorrências de cada valor único na coluna (agora contendo strings). Retorna uma série que mapeia cada valor único para sua frequência.

`len()`: Calcula o comprimento (número de elementos) da série retornada por value_counts(). Isso fornece o número de valores únicos na coluna após convertê-la para strings.


# **Segunda parte do FOR :**

2. `valores_unicos.append(len(df1[i].astype(str).value_counts()))`

`valores_unicos`: Destina-se a armazenar o número de valores únicos encontrados em cada coluna.

`.append()`: Adiciona um elemento ao final da lista valores_unicos.
O valor anexado é o mesmo resultado calculado na linha anterior. Isso armazena o número de valores únicos para cada coluna na lista valores_unicos.






```python
#Verificar o total de valores únicos de cada variável

valor_unico = []

for i in df1.columns[0:20].tolist():
  print(i, ':', len(df1[i].astype(str).value_counts()))
  valor_unico.append(len(df1[i].astype(str).value_counts()))
```
